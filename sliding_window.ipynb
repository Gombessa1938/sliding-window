```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import Rectangle
array = plt.imread('/content/Fits0.jpg')
```

```python
def sliding_window(image, stepSize, windowSize):

	for y in range(0, image.shape[0], stepSize):
		for x in range(0, image.shape[1], stepSize):
			
			yield x, y, image[y:y - windowSize[1], x:x + windowSize[0]]
```

```python
winW, winH = 200, 200

step = 100

plt.figure(figsize=(10,10))
plt.imshow(array)
counter = 0
for x, y, window in sliding_window(array, stepSize=step, windowSize=(winW, winH)):
  if window.shape[0] != winH or window.shape[1] != winW:
    counter += 1


    # note ! before you feed into CNN, change dtype to float32 ! very important
    # then 1/255 normalize 

    #here I m just drawing a rectangle, you want to put detection function here 
    #if detection(sliced image) ==1 : then draw rectange: 

    
    plt.gca().add_patch(Rectangle((x,y),winW,winH,
                  edgecolor='green',
                  facecolor='none',
                  lw=3))
    
    plt.annotate(counter,(x,y))


plt.show()
```
